```python
import re
import pandas as pd
```

```python
f = open('WhatsApp Chat with Krish.txt','r',encoding='utf-8')
```

```python
data = f.read()
```

```python
print(data)
```

```python
pattern = r'\d{1,2}/\d{1,2}/\d{2},\s\d{1,2}:\d{2}\s[APap][Mm]\s-\s'
```

```python
messages = re.split(pattern, data)[1:]
messages
```

```python
dates = re.findall(pattern, data)
dates = [s.replace('\u202F', ' ') for s in dates]
dates
```

```python
df = pd.DataFrame({'user_message':messages, 'date':dates})
df['date'] = pd.to_datetime(df['date'], format='%m/%d/%y, %I:%M %p - ')
df.head()
```

```python
df.shape
```

```python
users = []
messages = []
for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s', message)
    if entry[1:]:
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group notification')
        messages.append(entry[0])

df['user'] = users
df['message'] = messages
df.drop(columns=['user_message'], inplace=True)
df.head()
```

```python
df['year'] = df['date'].dt.year
df['month']= df['date'].dt.month_name()
df['day'] = df['date'].dt.day
df['hour'] = df['date'].dt.hour
df['minute'] = df['date'].dt.minute
df.head()
```

```python
df['month_num'] = df['date'].dt.month
timeline = df.groupby(['year','month_num','month']).count()['message'].reset_index()
```

```python
time=[]
for i in range(timeline.shape[0]):
    time.append(timeline['month'][i]+"-"+str(timeline['year'][i]))
```

```python
timeline['time']=time
```

```python
import matplotlib.pyplot as plt
plt.plot(timeline['time'],timeline['message'])
plt.xticks(rotation='vertical')
plt.show()
```

```python
df['date_only'] = df['date'].dt.date
daily_timeline = df.groupby('date_only').count()['message'].reset_index()
plt.figure(figsize=(18,10))
plt.plot(daily_timeline['date_only'],timeline['message'])
```
